# Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## loding data from seaborn for cars crashes

```python
cars = sns.load_dataset('car_crashes')
```

## display first five raws of data set

```python
cars.head()
```

```python
sns.relplot(x='total', y='speeding', data=cars);
```

```python
total_sums = cars.groupby('total', as_index=False).sum()
total_mean = cars.groupby('total', as_index=False).mean()

```

## display the summation of data 

```python
sns.relplot(x='total', y='speeding', data=total_sums);
```

## display the mean of data

```python
sns.relplot(x='total', y='speeding', data=total_mean);
```

```python
sns.lmplot(x="total", y="speeding", data=total_mean);
```

```python
sns.lmplot(x="total", y="speeding", data=total_sums,line_kws={'color': 'red'});
```

```python
graph = sns.barplot(x="total", y="speeding", data=cars ,ci=None)
```
